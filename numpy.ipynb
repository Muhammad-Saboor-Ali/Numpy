# Type Casting in NumPy

```python
import numpy as np

arr = np.array([1, 2, 3])

print(arr.dtype)
```

```python
arr = arr.astype(float)

print(arr)
print(arr.dtype)
```

```python
arr = np.array([1.5, 2.9, 3.2])

arr = arr.astype(int)

print(arr)
```

# Universal Functions (ufuncs)

```python
arr = np.array([1,4,9,16])
```

```python
np.sqrt(arr)
```

```python
np.exp(arr)
```

```python
np.sin(arr)
```

```python
np.abs(np.array([-1,-2,3]))
```
